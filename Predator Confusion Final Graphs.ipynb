# Predator Attack Efficiency as a Function of Nearby and Visible Prey

$P_{\rm{kill}} = \frac{1}{A_{\rm{NV}}}$

$A_{\rm{NV}}$ is the number of prey agents that are visible to the predator (i.e., in the predator agent's visual field) *and* are near the target prey (within 30 units; this includes the target prey itself)

```python
from pandas import *
import glob, numpy as np

figure(figsize=(7, 3.75))
ax = axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
xlabel("Number of Prey", fontsize="12")
ylabel("Predator Attack Efficiency", fontsize="12")
xticks(range(251)[::25])
yticks(arange(0, 1.01, 0.1))
xlim(0, 275)
ylim(0, 1)

dataLists = {}

# read data
for aeFile in glob.glob("coevolution-of-swarming/attack-efficiency-over-swarm-size/*"):
    dataLists[int(aeFile.split("/")[2].split(".")[0].split("swarmsize")[1])] = read_csv(aeFile)

# organize Attack Efficiency data for plotting
xList = []
yList = []
stdErrList = []

for SwarmSize in sorted(dataLists.keys()):
    NumberAttacks = dataLists[SwarmSize]["num_attacks"]
    NumberKilled = (float(SwarmSize) - dataLists[SwarmSize]["num_alive_end"])
    MeanAttackEfficiency = (NumberKilled / NumberAttacks).mean()
    StdErrAttackEfficiency = 2.0 * (NumberKilled / NumberAttacks).std() / sqrt(len(NumberKilled))

    xList.append(int(SwarmSize))
    yList.append(MeanAttackEfficiency)
    stdErrList.append(StdErrAttackEfficiency)
    
errorbar(x = xList, y = yList, yerr = stdErrList, fmt="k--", label="Total Swarm Size")
#savefig("coevolution-of-swarming/final-graphs/observed-attack-efficiency-total-swarm.pdf")

A_NV = [1, 2, 3, 4, 5, 10] + xList #range(1, 251)

AttackEfficiency = []

for Density in A_NV:
    AttackEfficiency.append(1.0 / float(Density))

plot(A_NV, AttackEfficiency, "^k-", label="A$_{NV}$")

legend(prop={'size':12})

"""figure(figsize=(7, 3.75))
plot(AttackEfficiency, "k-", linewidth=3)
xlabel("# of Nearby and Visible Prey", fontsize="12")
ylabel("Predator Attack Efficiency", fontsize="12")
ax = axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
xticks(fontsize="12")
yticks(fontsize="12")"""

savefig("coevolution-of-swarming/final-graphs/predator-attack-efficiency.pdf")
```

# Swarm Density, Swarm Dispersion, and Prey Agent Survivorship

```python
from pandas import *
import glob
import numpy as np

dataLists = {}

# read data
for folder in (glob.glob("coevolution-of-swarming/swarm-sd30") +
                glob.glob("coevolution-of-swarming/swarm-sd30-nopredconfusion")):
    
    dataLists[folder.split("/")[1]] = []
    
    for datafile in glob.glob(folder + "/*.csv"):
        
        replicateDF = read_csv(datafile)
        
        # sqrt the avg shortest distance because the simulation works in distance squared
        replicateDF["avg_shortest_dist"] = replicateDF["avg_shortest_dist"].apply(np.sqrt)
        
        dataLists[folder.split("/")[1]].append(replicateDF)

# calculate stats for data
meanDFs = {}
stderrDFs = {}

for key in dataLists.keys():
    
    keyDF = (concat(dataLists[key], axis=1, keys=range(len(dataLists[key])))
            .swaplevel(0, 1, axis=1)
            .sortlevel(axis=1)
            .groupby(level=0, axis=1))
    
    meanDFs[key] = keyDF.mean()
    stderrDFs[key] = keyDF.std().div(sqrt(len(dataLists[key]))).mul(2.0)
    
    keyDF = None

# plot data
for column in meanDFs[key].columns:
    
    #keyList = sorted(meanDFs.keys())
    keyList = ["swarm-sd30", "swarm-sd30-nopredconfusion"]
    
    if (column == "avg_shortest_dist") or (column == "swarm_density_count") or (column == "num_alive_end"):
        
        figure(figsize=(7, 3.75), frameon=False)
        
        if column == "avg_shortest_dist":
            ylabel("Mean Shortest Distance (units)", fontsize="12")
            ylim(ymax=70)
        elif column == "swarm_density_count":
            ylabel("Mean Number of Nearby Prey", fontsize="12")
            ylim(ymax=15)
        elif column == "num_alive_end":
            ylabel("Mean Surviving Prey", fontsize="12")
            ylim(ymax=50)
            xlabel("Generation", fontsize="12")
        
        ax = axes()
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()
        
        xticks(np.arange(0, 1201, 100), fontsize="12")
        yticks(fontsize="12")
        xlim(-20, 1220)
        
        for key in keyList:
            
            keyLabel = None
            
            if key == "swarm-sd30-nopredconfusion":
                keyLabel = "Without predator confusion"
                barFormat = "^--"
                barColor = "0.75"
            elif key == "swarm-sd30":
                keyLabel = "With predator confusion"
                barFormat = "o-"
                barColor = "0.0"
            
            errorbar(x=meanDFs[key]["generation"][::100], y=meanDFs[key][column][::100], yerr=stderrDFs[key][column][::100], label=keyLabel, fmt=barFormat, color=barColor, ms=10)
            
        legend(loc=2, prop={'size':12})
        savefig("coevolution-of-swarming/final-graphs/" + column.replace("_", "-") + "-LOD.pdf")
```

```python
from pandas import *
import glob
import numpy as np

dataLists = {}

# read data
for folder in glob.glob("coevolution-of-swarming/swarm-sd*-confusion-multiplier1"):
    
    dataLists[folder.split("/")[1]] = []
    
    for datafile in glob.glob(folder + "/*.csv"):
        
        replicateDF = read_csv(datafile)
        
        # sqrt the avg shortest distance because the simulation works in distance squared
        replicateDF["avg_shortest_dist"] = replicateDF["avg_shortest_dist"].apply(np.sqrt)
        
        dataLists[folder.split("/")[1]].append(replicateDF)

# calculate stats for data
meanDFs = {}
stderrDFs = {}

for key in dataLists.keys():
    
    keyDF = (concat(dataLists[key], axis=1, keys=range(len(dataLists[key])))
            .swaplevel(0, 1, axis=1)
            .sortlevel(axis=1)
            .groupby(level=0, axis=1))
    
    meanDFs[key] = keyDF.mean()
    stderrDFs[key] = keyDF.std().div(sqrt(len(dataLists[key]))).mul(2.0)
    
    keyDF = None

# plot data
for column in meanDFs[key].columns:
    
    #keyList = sorted(meanDFs.keys())
    keyList = ["swarm-sd0-confusion-multiplier1", "swarm-sd30-confusion-multiplier1"]
    
    if (column == "swarm_density_count") or (column == "num_attacks"):
        
        figure(figsize=(7, 3.75), frameon=False)
        
        if column == "swarm_density_count":
            ylabel("Mean Number of Nearby Prey", fontsize="12")
            ylim(-1, 25)
        elif column == "num_attacks":
            ylabel("Number of Attacks by Predator", fontsize="12")
            xlabel("Generation", fontsize="12")
        
        ax = axes()
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()
        
        xticks(np.arange(0, 1201, 100), fontsize="12")
        yticks(fontsize="12")
        xlim(-20, 1220)
        
        for key in keyList:
            
            keyLabel = None
            
            if key == "swarm-sd0-confusion-multiplier1":
                keyLabel = "Without predator confusion"
                barFormat = "^--"
                barColor = "0.75"
            elif key == "swarm-sd30-confusion-multiplier1":
                keyLabel = "With predator confusion"
                barFormat = "o-"
                barColor = "0.0"
            
            errorbar(x=meanDFs[key]["generation"][::100], y=meanDFs[key][column][::100], yerr=stderrDFs[key][column][::100], label=keyLabel, fmt=barFormat, color=barColor, ms=10)
            
        legend(loc=2, prop={'size':12})
        savefig("coevolution-of-swarming/final-graphs/" + column.replace("_", "-") + "-grouping-cost.pdf")
```

# Predator attack efficiency as a function of nearby and visible prey

```python
A_NV = range(1, 51)

AttackEfficiency1N = []
AttackEfficiency15N = []
AttackEfficiency2N = []
AttackEfficiency5N = []

for Density in A_NV:
    AttackEfficiency1N.append(1.0 / float(Density))
    AttackEfficiency15N.append(min(1.5 / float(Density), 1.0))
    AttackEfficiency2N.append(min(2.0 / float(Density), 1.0))
    AttackEfficiency5N.append(min(5.0 / float(Density), 1.0))

figure(figsize=(7, 3.75))
plot(AttackEfficiency1N, "k-", linewidth=2, label="1 / N")
plot(AttackEfficiency15N, "k--", linewidth=2, label="1.5 / N")
plot(AttackEfficiency2N, "k-.", linewidth=2, label="2 / N")
plot(AttackEfficiency5N, "k:", linewidth=2, label="5 / N")
xlabel("# of Nearby and Visible Prey", fontsize="12")
ylabel("Predator Attack Efficiency", fontsize="12")
legend(prop={'size':12})
ax = axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
xticks(fontsize="12")
yticks(fontsize="12")
savefig("coevolution-of-swarming/final-graphs/variable-predator-attack-efficiency.pdf")
```

# Observed predator attack efficiency as a function of total swarm size

```python
from pandas import *
import glob, numpy as np

figure(figsize=(7, 3.75))
ax = axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
xlabel("Swarm Size", fontsize="12")
ylabel("Predator Attack Efficiency", fontsize="12")
xticks(range(251)[::25])
yticks(arange(0, 1.01, 0.1))
xlim(0, 275)
ylim(0, 1)

dataLists = {}

# read data
for aeFile in glob.glob("coevolution-of-swarming/attack-efficiency-over-swarm-size/*"):
    dataLists[int(aeFile.split("/")[2].split(".")[0].split("swarmsize")[1])] = read_csv(aeFile)

# organize Attack Efficiency data for plotting
xList = []
yList = []
stdErrList = []

for SwarmSize in sorted(dataLists.keys()):
    NumberAttacks = dataLists[SwarmSize]["num_attacks"]
    NumberKilled = (float(SwarmSize) - dataLists[SwarmSize]["num_alive_end"])
    MeanAttackEfficiency = (NumberKilled / NumberAttacks).mean()
    StdErrAttackEfficiency = 2.0 * (NumberKilled / NumberAttacks).std() / sqrt(len(NumberKilled))

    xList.append(int(SwarmSize))
    yList.append(MeanAttackEfficiency)
    stdErrList.append(StdErrAttackEfficiency)
    
errorbar(x = xList, y = yList, yerr = stdErrList, fmt="k-")
savefig("coevolution-of-swarming/final-graphs/observed-attack-efficiency-total-swarm.pdf")
```
